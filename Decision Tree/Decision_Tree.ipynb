## **Decision Tree**

Decision Trees are versatile machine learning algorithms.<br><br>
**They can handle:**
- Classification
- Regression
- Multioutput tasks

**Advantages:**
- Powerful and capable of fitting complex datasets.
- Easy to understand and visualize.
- Serve as the foundation for Random Forests, one of the most powerful ML algorithms.

### **Training a Decision Tree**

**iris Dataset** <br><br>

- **Sepal length:** the length of the sepal in centimeters
- **Sepal width:** the width of the sepal in centimeters
- **Petal length:** the length of the petal in centimeters
- **Petal width:** the width of the petal in centimeters
- **Species:** the species of the iris flower (Setosa, Versicolor, or Virginica)

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris() #reading iris dataset
```

```python
iris.data #features of iris dataset
```

```python

#0-> setosa
#1-> versicolor
#2-> virginica

iris.target #label of iris dataset
```

```python
X=iris.data[:, 2:] #petal length and width
y=iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

#### **Visualize Decision Tree**

Text View

```python
from sklearn.tree import export_text
```

```python
text=export_text(tree_clf, feature_names=iris.feature_names[2:], class_names=iris.target_names)
print(text)
```

```python
from sklearn.tree import plot_tree
```

```python
tree=plot_tree(tree_clf, feature_names=iris.feature_names[2:], class_names=iris.target_names)
print(tree)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

On scatter plot

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = tree_clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Reshape the predictions to match the grid shape
Z = Z.reshape(xx.shape)

# Plot the decision boundaries
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)

# Plot the dataset
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k", cmap="viridis")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Decision Tree Splitting Visualized")
plt.legend(*scatter.legend_elements(), title="Classes")
plt.show()
```

#### **How Decision Trees Make Predictions:**
- Start at the root node:<br>
    o Check if the condition at the node is true or false.<br>
    o Traverse to the corresponding child node.<br>
    <br>
**Example:**<br>
    - If petal length < 2.45 cm → Move to left child (predict Iris-Setosa).
    - If petal length > 2.45 cm → Move to right child and check:
    - If petal width < 1.75 cm → Predict Iris-Versicolor.
    - Otherwise → Predict Iris-Virginica.

#### **Decision Boundaries:**
- **Represented by:**
    - Vertical, dashed, and dotted lines depending on node depth.
- **Example:**
    - Root node (Depth 0): Vertical line at petal length = 2.45 cm.
    - Right child (Depth 1): Dashed line at petal width = 1.75 cm.

#### **Model Interpretability:**
- **Decision Trees:**
    - **White-box models:** Easy to understand and explain.
- **Random Forests/Neural Networks:**
    - **Black-box models:** Good predictions but hard to explain the logic behind them.

#### **CART Algorithm**

**Basic Idea:**
- CART (Classification And Regression Tree) splits the dataset into subsets based on a feature 𝑘 and a threshold 𝑡𝑘 (e.g., petal length ≤ 2.45 cm).
- The algorithm selects the split that minimizes impurity, measured by a cost function:


**How the Tree Grows:**
- Recursively splits subsets until:
    - Maximum depth (max_depth) is reached.
    - Further splits don’t reduce impurity.

**Stopping Conditions:**
- Controlled by hyperparameters like:
    - min_samples_split (minimum samples required to split a node).
    - min_samples_leaf (minimum samples in a leaf node).
    - max_leaf_nodes (maximum number of leaf nodes).

**Greedy Algorithm:**
- Optimizes splits at each level without considering future levels.
- Produces a good (but not optimal) tree due to computational constraints.

**Why Not Optimal Trees?**
- Finding the best tree is NP-Complete (computationally intractable for large datasets).
- Approximate solutions are used.

#### **Gini Impurity vs. Entropy**

- **Gini Impurity (default):**
    - Measures impurity.
    - Faster to compute.
    - Tends to isolate the most frequent class in its own branch.

- **Entropy:**
    - Based on Shannon’s information theory.
    - Produces slightly more balanced trees.
    - Computationally heavier.

- **Key Insight:**
    - Both often lead to similar results. Use Gini as the default unless balanced splits are critical.

#### **Regularization in Decision Trees**

- **Nonparametric Nature of Decision Trees:**
    - **Nonparametric models:** The structure is not fixed; it adapts to the training data, which increases the risk of overfitting.
    - **Parametric models (e.g., linear models):** Fixed number of parameters, reducing the risk of overfitting but potentially leading to underfitting.

- **Regularization in Decision Trees:**
    - Regularization limits the tree's freedom to reduce overfitting by restricting its structure during training.
    - **Key Hyperparameters for Regularization:**
        - **max_depth:** Limits the depth of the tree (default is None for no limit).
        - **min_samples_split:** Minimum number of samples required to split an internal node.
        - **min_samples_leaf:** Minimum number of samples a leaf node must have.
        - **min_weight_fraction_leaf:** Similar to min_samples_leaf but based on a fraction of the total weighted instances.
        - **max_leaf_nodes:** Limits the number of leaf nodes.
        - **max_features:** Limits the number of features considered for splitting at each node.

- **Effects of Regularization:**
    - Increasing min_* hyperparameters or decreasing max_* hyperparameters regularizes the model:
        - Reduces overfitting.
        - Potentially increases generalization.

- **Pruning:**
    - Some algorithms use pruning to regularize the model:
        - Train the tree without restrictions first.
        - Remove unnecessary nodes whose children are leaf nodes, provided the purity improvement is statistically insignificant.
        - Use statistical tests like the χ² test to decide whether a node's improvement is due to chance.
        - If the p-value exceeds a threshold (typically 5%), the node is pruned.

- **Example Scenario:**
    - A Decision Tree trained with default hyperparameters (unrestricted) overfits, producing a complex structure that matches the training data too closely.
    - A Decision Tree with min_samples_leaf=4 creates a more generalized model with better predictions on unseen data.
